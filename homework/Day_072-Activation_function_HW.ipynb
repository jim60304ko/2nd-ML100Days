# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
x=plt.linspace(-10,10,100)
def ReLU(x):
    return (abs(x)*(x>0))
def dReLU(x):
    return (1*(x>0))
plt.plot(x,ReLU(x),'r')
plt.plot(x,dReLU(x),'b')
plt.title('ReLU and dReLU')
```
